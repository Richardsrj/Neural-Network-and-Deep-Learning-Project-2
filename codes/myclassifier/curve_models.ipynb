```python
import matplotlib.pyplot as plt
import numpy as np


# 文件名和标签
files = {
    "adamW_weight_decay1.txt": "basic_model",
    "adamW_large.txt": "large_model",
    "adamW_weight_light.txt": "light_model"
}



metric_idx = 0  # 比较 Validation Accuracy

plt.figure(figsize=(10, 6))

for filename, label in files.items():
    data = np.loadtxt(filename, comments='#')
    epochs = np.arange(1, len(data)+1)
    plt.plot(epochs, data[:, metric_idx], label=label)

plt.xlabel("Epoch")
plt.ylabel("Train Loss")
plt.title("Activation Function Comparison--train loss")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("activation_comparison.png")
plt.show()
```

```python
metrics = ['Train Loss', 'Validation Loss', 'Train Accuracy', 'Validation Accuracy']
ylabels = ['Train Loss', 'Val Loss', 'Train Acc', 'Val Acc']
titles = ['Activation Function Comparison -- Train Loss',
          'Activation Function Comparison -- Val Loss',
          'Activation Function Comparison -- Train Accuracy',
          'Activation Function Comparison -- Val Accuracy']

fig, axes = plt.subplots(2, 2, figsize=(14, 10))  # 2行2列子图
axes = axes.flatten()  # 展平方便索引

for metric_idx in range(4):
    ax = axes[metric_idx]
    for filename, label in files.items():
        data = np.loadtxt(filename, comments='#')
        epochs = np.arange(1, len(data)+1)
        ax.plot(epochs, data[:, metric_idx], label=label)
    
    ax.set_title(titles[metric_idx])
    ax.set_xlabel('Epoch')
    ax.set_ylabel(ylabels[metric_idx])
    ax.grid(True)
    ax.legend()

plt.tight_layout()
plt.savefig("activation_comparison_all.png")
plt.show()
```
